```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# Esto carga las variables del archivo .env en la memoria del sistema
load_dotenv()

# Ahora el cliente puede tomar la clave automáticamente sin problemas
client = OpenAI()

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "¡Hola! Estoy probando mi entorno del Zoomcamp."}]
)

print(response.choices[0].message.content)
```

```python
# Definimos la función wrapper para el LLM (Nuestra caja negra)
def llm(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

# Probamos que la función responda (El calentamiento)
print(llm("Hey, what's up?"))
```

```python
# Hacemos la pregunta de reglamento del Zoomcamp
query = "I just discovered the course. Can I join now?"

# Se la mandamos al modelo sin contexto previo
answer = llm(query)
print(answer)
```

```python
# 1. Definimos la información oficial que recuperamos del vestuario (El Contexto)
context = """
I just discovered the course. Can I still join?
Yes, but if you want to receive a certificate, you need to submit your project while we're still accepting submissions.

Course: I have registered for the LLM Zoomcamp. When can I expect to receive the confirmation email?
You don't need it. You're accepted. You can also just start learning and submitting homework (while the form is open) without registering. It is not checked against any registered list. Registration is just to gauge interest before the start date.

What is the video/zoom link to the stream for the "Office Hours" or live/workshop sessions?
The zoom link is only published to instructors/presenters/TAs. Students participate via YouTube Live and submit questions to Slido.
"""

# 2. Armamos la indicación táctica (Inyectamos la pregunta y el contexto)
prompt = f"""
Your task is to answer questions from the course participants based on the provided context.
Use the context to find relevant information and provide accurate answers. 
If the answer is not found in the context, respond with "I don't know."

Question: {query}
Context: {context}
""".strip()

# 3. Mandamos la jugada preparada al LLM
print(llm(prompt))
```
